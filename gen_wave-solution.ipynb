
# 生成波形


- 利用 numpy 库生成采样率为 44100 ,长度为 2s 的方波和三角波
- 并使用 librosa 库将波形转化为音频


需要 numpy, librosa 库

```python
from __future__ import print_function

import librosa
import numpy as np
```

本次实验的使用的音频如下：

```python
# code 1
# 设定波形的频率值与幅度值
f = 200  # 波形频率值 
amp =  1 # 幅度值
sr = 44100  # 设定采样率
```

利用 numpy.zeros 初始化两个波形数组, pulse_wave 和 triangle_wave

```python
# code 2
num_cycle = 2 # 波形持续多少秒
pulse_wave = np.zeros((sr * num_cycle)) # 每秒采样 sr 个点 * 持续秒数
triangle_wave = np.zeros((sr * num_cycle))
```

根据方波的特性给 pulse_wave 每一个帧赋值
ps: 前半个周期值为设定的幅度值, 后半个周期值为0

```python
# code 3
for c_i in range(num_cycle): # 每秒
        for s_i in range(sr): # 每个采样点
            # 计算采样点的值
            # s_i % (sr / f)  确定是第几秒的波形
            #  后面的 / (sr / f)  确定是波形的哪个部分
            percent = s_i % (sr / f) / (sr / f)
            if percent < 0.5:
                amp_i = amp
            else:
                amp_i = 0
            pulse_wave[c_i*sr + s_i] = amp_i
```

根据三角波的特性给 triangle_wave 每一个帧赋值
ps: 前半个周期线性递增, 后半个周期线性递减

```python
# code 4
for c_i in range(num_cycle): # 每秒
        for s_i in range(sr): # 每个采样点
            # 计算采样点的值
            # s_i % (sr / f)  确定是第几秒的波形
            #  后面的 / (sr / f)  确定是波形的哪个部分
            percent = s_i % (sr / f) / (sr / f)
            if percent < 0.5:
                amp_i = amp * percent
            else:
                amp_i = amp * (1 - percent)
            triangle_wave[c_i*sr + s_i] = amp_i
```

将两个波形保存为音频文件

```python
# code 5
pulse_path = 'pulse.wav'
triangle_path = 'triangle.wav'
librosa.output.write_wav(pulse_path, pulse_wave, sr=sr)
librosa.output.write_wav(triangle_path, triangle_wave, sr=sr)
```

```python
import IPython.display as ipd
# 播放方波声音
ipd.Audio(pulse_path)
```

```python
# 播放三角波声音
ipd.Audio(triangle_path)
```
